```python
import pandas as pd
df = pd.read_csv("nlpp.csv")
print(df)
```

```python
df=df[df['pos']!='START']
df=df[df['pos']!='END']
```

```python
import numpy as np

tags = np.unique(df['pos'])
def funTwo(word):
  c=np.where(tags==word)
  return c[0][0]
df['index_pos']=df['pos'].apply(funTwo)
```

```python
df.head()
```

```python
voc=np.unique(df['word'])
voc[:10]
voc_size=len(voc)
voc_size
```

```python
tags
tags2id={i:w for i,w in enumerate(tags)}
tags2id
```

```python
sen=[]
tot_sen=[]
pos=[]
tot_pos=[]
ind=[]
tot_ind=[]

c=-1
for i in df.index:
  if c+1==i:
    sen.append(df['word'][i])
    pos.append(df['pos'][i])
    ind.append(df['index_pos'][i])
    c=i
  else:
    tot_sen.append(sen)
    tot_pos.append(pos)
    tot_ind.append(ind)
    pos=[]
    sen=[]
    ind=[]
    c=i
```

```python
update_df = pd.DataFrame({'sen':tot_sen,'pos':tot_pos,'index_pos':tot_ind})
update_df.head()
```

```python
corpus=[]
def dead(sen):
    sent=' '.join([str(elem) for elem in sen])
    corpus.append(sent)
    return sent
update_df['sent']=update_df['sen'].apply(dead)
```

```python
corpus
```

```python
update_df['sent'].to_csv('data.txt', index=False, sep='\t')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Dense, LSTM, Input, Dropout,Embedding,TimeDistributed, Bidirectional
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.text import one_hot
```

```python
word2id = {w: i for i, w in enumerate(df['word'])}
word2id
```

```python

onehot_rep=[one_hot(words,voc_size)for words in corpus]
onehot_rep
```

```python
padded_sen=pad_sequences(onehot_rep,padding="post",maxlen=30)
padded_sen
```

```python
padded_tag_pos=pad_sequences(tot_ind,padding="post",maxlen=30)
padded_tag_pos
```

```python
y = [to_categorical(i, num_classes=len(tags)) for i in padded_tag_pos]
y[0]
```

```python
X=padded_sen
y=y
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.20)
```

```python
print(y_tr[0][1])
print("-----------")
print(y_te[0][1])
print("-----------")
print(y[0][1])
```

```python
max_len=30
input = Input(shape=(max_len,)) # Input layer
model = Embedding(input_dim=7073, output_dim=30, input_length=max_len)(input) # Word embedding layer
model = Dropout(0.1)(model) # Dropout
model = Bidirectional(LSTM(units=100, return_sequences=True, recurrent_dropout=0.1))(model) # Bi-directional LSTM layer
out = TimeDistributed(Dense(len(tags), activation="softmax"))(model)  # softmax output layer
```

```python
model = Model(input, out)
```

```python
model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=1e-5), metrics=['accuracy'])
```

```python
history = model.fit(np.array(X), np.array(y), batch_size=32, epochs=64, validation_split=0.1, verbose=1)
```
